```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('weight-height.csv')
```

```python
df.head()
```

```python
x = df['Height']
y = df['Weight']
```

```python
## train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
##cross validation
from sklearn.model_selection import cross_val_score
```

```python
regression=LinearRegression()
regression.fit(x_train.values.reshape(-1, 1),y_train)
```

```python
LinearRegression()
```

```python
mse = cross_val_score(regression, x_train.values.reshape(-1, 1), y_train, scoring='neg_mean_squared_error', cv=100)
mean_mse = np.mean(mse)
print(mean_mse)
```

```python
## prediction
pred = regression.predict(x_test.values.reshape(-1, 1))
```

```python
## display the result
sns.distplot(y_test-pred, kde=True)
```

```python
## Scoring the model with r2
from sklearn.metrics import r2_score
```

```python
score = r2_score(y_test, pred)
print(score)
```
