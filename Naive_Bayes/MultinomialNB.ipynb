```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('spam_or_not_spam.csv')
df.head()
```

```python
X = df['email']
y = df['label']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
vectorizer = CountVectorizer()
X_train_vectors = vectorizer.fit_transform(X_train.fillna(''))
X_test_vectors = vectorizer.transform(X_test.fillna(''))
```

```python
model = MultinomialNB()
model.fit(X_train_vectors, y_train)
```

```python
y_pred = model.predict(X_test_vectors)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```

```python
custom_message = ["Congratulations, you've won a free vacation"]
custom_vector = vectorizer.transform(custom_message)
prediction = model.predict(custom_vector)
print("Spam" if prediction[0] == 1 else "Not Spam")
```
